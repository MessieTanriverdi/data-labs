```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/forks"
```

```python
def token():
    with open("/Users/messie/Desktop/GitHub/api.text", "r") as f: 
        lines = f.readlines() 
        for line in lines: 
            return(line)
        
```

```python
response = requests.get(url, headers={"Authorization": str(token())})

results = response.json()

results
```

```python
flattened_data = json_normalize(results)

df = flattened_data

flattened_data.head(2)
```

```python
# Sur la DF

set(df['language'])
```

```python
# Sur la liste

lang = []

for i in results:
    # i = dict
    l = i['language']
    if l not in lang:
        lang.append(l)
    
lang
```
